```python
# Definir a fun칞칚o f(x) = x^3
def f(x):
    return x**3

# Criar um intervalo de valores de x de 1 a 2
x = np.linspace(1, 2, 100)

# Calcular os valores de y usando a fun칞칚o f(x)
y = f(x)

# Criar o gr치fico
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$f(x) = x^3$', color='blue')
plt.fill_between(x, y, alpha=0.2, color='blue', step='post')

# Adicionar t칤tulo e r칩tulos aos eixos
plt.title('Gr치fico da fun칞칚o $f(x) = x^3$')
plt.xlabel('x')
plt.ylabel('f(x)')

# Adicionar uma legenda
plt.legend()

# Adicionar uma grade para melhor visualiza칞칚o
plt.grid(True)

# Mostrar o gr치fico
plt.show()
```

### Forma num칠rica da integral da fun칞칚o  洧논3:


```python
# Definir a fun칞칚o f(x) = x^3
def f(x):
    return x**3

# Definir o intervalo e o n칰mero de ret칙ngulos
a = 1
b = 2
n = 4  # N칰mero de ret칙ngulos
delta_x = (b - a) / n

# Calcular a soma das 치reas dos ret칙ngulos
integral_approx = 0
for i in range(n):
    x_i = a + i * delta_x
    integral_approx += f(x_i) * delta_x

print(f"Integral aproximada usando a regra dos ret칙ngulos com {n} ret칙ngulos: {integral_approx:.6f}")
```

### Forma analitica da integral da fun칞칚o  洧논3:

```python
import sympy as sp

# Definir a vari치vel simb칩lica
x = sp.Symbol('x')

# Definir a fun칞칚o f(x) = x^3
f = x**3

# Calcular a integral definida
a = 1
b = 2
integral_definida = sp.integrate(f, (x, a, b))

# Obter o valor num칠rico da integral
valor_numerico = integral_definida.evalf()

print(f"Integral definida de x^3 entre {a} e {b}: {valor_numerico:.6f}")
```

A diferen칞a entre as duas foi enorme, mas conforme aumentassemos o n칰mero de ret칙ngulos na n칰mero 1, mais pr칩ximo ficariam os resultados. Perceba, agora com 60 ret칙ngulos:

```python
# Definir a fun칞칚o f(x) = x^3
def f(x):
    return x**3

# Definir o intervalo e o n칰mero de ret칙ngulos
a = 1
b = 2
n = 60  # N칰mero de ret칙ngulos
delta_x = (b - a) / n

# Calcular a soma das 치reas dos ret칙ngulos
integral_approx = 0
for i in range(n):
    x_i = a + i * delta_x
    integral_approx += f(x_i) * delta_x

print(f"Integral aproximada usando a regra dos ret칙ngulos com {n} ret칙ngulos: {integral_approx:.6f}")
```

A diferen칞a com 60 ret칙ngulos 칠 muito menor do que com 4, infelizmente, no papel 칠 muito mais dif칤cil chegar nesse resultado visto que daria muito mais trabalho, o que torna a forma anal칤tica muito mais pr치tica.
