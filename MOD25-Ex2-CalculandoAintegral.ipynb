1- Calcule de forma numérica a integral da função $x^3$ avaliada entre os pontos $1$ e $2$. Identifique o número de retângulos que você usou.  
2- Calcule essa área de forma analítica.

```python
# Definir a função f(x) = x^3
def f(x):
    return x**3

# Criar um intervalo de valores de x de 1 a 2
x = np.linspace(1, 2, 100)

# Calcular os valores de y usando a função f(x)
y = f(x)

# Criar o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$f(x) = x^3$', color='blue')
plt.fill_between(x, y, alpha=0.2, color='blue', step='post')

# Adicionar título e rótulos aos eixos
plt.title('Gráfico da função $f(x) = x^3$')
plt.xlabel('x')
plt.ylabel('f(x)')

# Adicionar uma legenda
plt.legend()

# Adicionar uma grade para melhor visualização
plt.grid(True)

# Mostrar o gráfico
plt.show()
```

### Forma numérica da integral da função  𝑥3:


```python
# Definir a função f(x) = x^3
def f(x):
    return x**3

# Definir o intervalo e o número de retângulos
a = 1
b = 2
n = 4  # Número de retângulos
delta_x = (b - a) / n

# Calcular a soma das áreas dos retângulos
integral_approx = 0
for i in range(n):
    x_i = a + i * delta_x
    integral_approx += f(x_i) * delta_x

print(f"Integral aproximada usando a regra dos retângulos com {n} retângulos: {integral_approx:.6f}")
```

### Forma analitica da integral da função  𝑥3:

```python
import sympy as sp

# Definir a variável simbólica
x = sp.Symbol('x')

# Definir a função f(x) = x^3
f = x**3

# Calcular a integral definida
a = 1
b = 2
integral_definida = sp.integrate(f, (x, a, b))

# Obter o valor numérico da integral
valor_numerico = integral_definida.evalf()

print(f"Integral definida de x^3 entre {a} e {b}: {valor_numerico:.6f}")
```

A diferença entre as duas foi enorme, mas conforme aumentassemos o número de retângulos na número 1, mais próximo ficariam os resultados. Perceba, agora com 60 retângulos:

```python
# Definir a função f(x) = x^3
def f(x):
    return x**3

# Definir o intervalo e o número de retângulos
a = 1
b = 2
n = 60  # Número de retângulos
delta_x = (b - a) / n

# Calcular a soma das áreas dos retângulos
integral_approx = 0
for i in range(n):
    x_i = a + i * delta_x
    integral_approx += f(x_i) * delta_x

print(f"Integral aproximada usando a regra dos retângulos com {n} retângulos: {integral_approx:.6f}")
```

A diferença com 60 retângulos é muito menor do que com 4, infelizmente, no papel é muito mais difícil chegar nesse resultado visto que daria muito mais trabalho, o que torna a forma analítica muito mais prática.
