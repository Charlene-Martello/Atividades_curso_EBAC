####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2, 2) #2 linhas e 2 colunas, totaL: 4 gráficos

sns.countplot(ax = axes[0, 0], x='SEXO', data=nascidos_vivos) #primeiro gráfico
sns.countplot(ax = axes[0, 1], x='GESTACAO', data=nascidos_vivos) #2º
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=nascidos_vivos) #3º
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=nascidos_vivos)#4º

#ajustar o espaçamento
plt.subplots_adjust(wspace=1, hspace=1)
```
