# Cálculo - Tarefa 01

1. Marque quais desses métodos/algoritmos muito populares em ciência de dados são baseados no uso de derivada:

    1. Método Mínimos Quadrados
    2. Gradiente descendente
    3. Newton Raphson
    4. CART (Árvore de decisão)

* Método dos Mínimos Quadrados: Utiliza derivadas para encontrar os valores dos parâmetros que minimizam a função de erro.
* Gradiente descendente: Utiliza derivadas (ou gradientes) para iterativamente ajustar os parâmetros do modelo e minimizar uma função de custo. 
* Newton Raphson: Utiliza derivadas para encontrar as raízes de uma função ou otimizar uma função de custo.
* CART: Não utiliza derivadas, pois utiliza uma abordagem de divisão que não requer cálculos de gradientes.

2. Dada uma base de dados com uma variável resposta $y$ e um conjunto de variáveis explicativas. Considere uma estrutura de um modelo de regressão. Explique com suas palavras por que não é possível obter parâmetros que forneçam um erro quadrático médio (EQM) menor que o obtido com estimadores de mínimos quadrados.

O Erro Quadrático Médio (EQM ou MSE - Mean Squared Error) é uma métrica utilizada na estatística para avaliar a precisão de um modelo. Ele calcula a média dos quadrados das diferenças entre os valores previstos e os valores reais. O método dos Mínimos Quadrados, por sua vez, é uma técnica que visa minimizar essas diferenças ao ajustar um modelo aos dados. Assim, o MSE obtido com estimadores de Mínimos Quadrados é menor que o EQM, uma vez que é projetado para reduzí-lo.
